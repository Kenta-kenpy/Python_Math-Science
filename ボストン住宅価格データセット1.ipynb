## ボストン住宅価格データセット

ボストン住宅価格データセットインポート

```python
from sklearn import datasets
boston = datasets.load_boston() # 変数bostonにデータセットを格納
print(boston.keys())
```

データセットの概要を見る

```python
print(boston.DESCR)
```

Pandasデータフレームに格納して表にする

```python
import pandas as pd
df = pd.DataFrame(boston.data, columns = boston.feature_names)

df
```

住宅価格MEDV列を追加

```python
import pandas as pd

df = pd.DataFrame(boston.data, columns = boston.feature_names)

df["MEDV"] = boston.target

df
```

## グラフにして様々なデータの可視化

```python
import seaborn as sns
%matplotlib inline
```

### 犯罪率CRIM - 住宅価格MEDV

```python
sns.jointplot(x="CRIM", y="MEDV", data=df)
```

犯罪率はほとんどの場合で低いですが、犯罪率が高いのは低価格の住宅価格のところに偏っている。

この２つには関連性が認められる。

### 25000平方フィート以上の住宅区画割合ZN - 住宅価格MEDV

```python
sns.jointplot(x="ZN", y="MEDV", data=df)
```

```python
# 関連性がありそうなので回帰直線を描く
sns.jointplot(x="ZN", y="MEDV", data=df, kind="reg")
```

25000平方フィート以上の住宅区画割合と住宅価格も、なんとなく関連性がありそうだ。

### 小売業以外の割合INDUS - 住宅価格MEDV

```python
sns.jointplot(x="INDUS", y="MEDV", data=df)
```

小売業の割合と住宅価格も、なんとなく関連性がありそうだ。

### チャールズ川沿いCHAS - 住宅価格MEDV
